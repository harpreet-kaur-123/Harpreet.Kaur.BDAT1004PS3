```python
# step 1
from pandas import read_table
from datetime import datetime
import pandas as pd
import numpy as np
```

```python
# step 2 and 3
data = read_table('wind.txt', sep = '\s+')
#data = pd.read_csv("wind.txt")
data["Date"] = pd.to_datetime(data[["Yr","Mo","Dy"]].astype(str).agg('-'.join, axis=1))
data = data.drop(columns=["Yr","Mo","Dy"])
data.head()
```

```python
#step 4
data["Date"] = np.where(pd.DatetimeIndex(data["Date"]).year < 2000,data.Date,data.Date - pd.offsets.DateOffset(years=100))
```

```python
#step 5
data_df = data.set_index("Date")
data_df.index.astype("datetime64[ns]")
```

```python
#step 9
def statistics(x):
    x = pd.Series(x)
    Min = x.min()
    Max = x.max()
    Mean = x.mean()
    Std = x.std()
    res = [Min,Max,Mean,Std]
    indx = ["Min","Max","Mean","Std"]
    res = pd.Series(res,index=indx)
    return res
loc_stats = data_df.apply(statistics)
loc_stats
```

# Question 6
Create a line plot showing the number of marriages and divorces per capita in the 
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!


```python
import pandas as pd
import pylab
import matplotlib.pyplot as plt

us_marriage_divorce_data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
years = us_marriage_divorce_data['Year'].values
marriages_per_capita = us_marriage_divorce_data['Marriages_per_1000'].values
divorces_per_capita  = us_marriage_divorce_data['Divorces_per_1000'].values

line1 = pylab.plot(years, marriages_per_capita, 'b-o',label='Marriage rate')
pylab.ylabel('Marriages per capital')

pylab.twinx()
line2 = pylab.plot(years, divorces_per_capita, 'r-o',label='Divorce rate')
pylab.ylabel('Divorces per capital')

lines = line1 + line2
labels = []
for line in lines:
    labels.append(line.get_label())

pylab.legend(lines, labels)
pylab.show()
```

# Question 7
Create a vertical bar chart comparing the number of marriages and divorces per 
capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

us_marriage_divorce_data = us_marriage_divorce_data[
us_marriage_divorce_data['Year'].apply(lambda x: x in [1900, 1950, 2000])]

years = us_marriage_divorce_data['Year'].values
marriages_per_capita = us_marriage_divorce_data['Marriages_per_1000'].values
divorces_per_capita = us_marriage_divorce_data['Divorces_per_1000'].values

X_axis = np.arange(len(years))
plt.bar(X_axis - 0.2, marriages_per_capita, 0.4, label = 'Marriages per capita')
plt.bar(X_axis + 0.2, divorces_per_capita, 0.4, label = 'Divorces per capita')
plt.xticks(X_axis, years)
plt.xlabel("Years")
plt.ylabel("Marriages and divorces per capita")
plt.title("Marriages and divorces per capita in the U.S. between 1900, 1950, and 2000")
plt.legend()
plt.show()
```

# Question 9
Create a pie chart showing the fraction of all Roman Emperors that were 
assassinated.
Make sure that the pie chart is an even circle, labels the categories, and shows the 
percentage breakdown of the categories.

```python
import pandas as pd
import matplotlib.pyplot as plt

roman_emperors = pd.read_csv('roman-emperor-reigns.csv')
assassinated_emperors = roman_emperors[roman_emperors['Cause_of_Death'].apply(lambda x: 'assassinated' in x.lower())]

number_assassinated = len(assassinated_emperors)
other_deaths = len(roman_emperors) - number_assassinated

names = ['Assassinated', 'Other deaths']
data = [number_assassinated, other_deaths]
fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels = names)
plt.show()
```
